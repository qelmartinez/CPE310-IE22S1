### `**LISTS:**`

*Create a List*

```python
# 1. Create a List
fruits = ["apple", "banana", "cherry", "orange"]
print("Original List: ", fruits)
```

*Access Elements*

```python
# 2. Access Elements
print("First Fruit: ", fruits[0])
print("Last Fruit: ", fruits[-1])     #pabaliktad so if [-2], it'll be: "Last Fruit: cherry"
```

*Add Elements*

```python
# 3. Add Elements
fruits.append("mango")      #Add at the end
fruits.insert(1, "grape")   #Add at specific position
print("After adding fruits: ", fruits)
```

*Remove Elements*

```python
# 4. Remove Elements
fruits.remove("banana")     #Remove by value
popped = fruits.pop()       #Remove last item
print("After removing fruits: ", fruits)
print("Popped Fruit: ", popped)
```

Check if an Item Exists

```python
# 5. Check if an Item Exists
if "apple" in fruits:
    print("Apple is in the list")
```

*Loop Through the List*

```python
# 6. Loop Through the List
print("All fruits in the list:")
for fruit in fruits:
    print("-", fruit)
```

*Length of the List*

```python
# 7. Length of the List
print("Total number of fruits: ", len(fruits))
```

*Sorting the List*

```python
# 8. Sorting the list
fruits.sort()     #Alphabetical order
print("Sorted fruits: ", fruits)
```

*Reversing the List*


```python
# 9. Reversing the list
fruits.reverse()
print("Reverse List: ", fruits)
```
