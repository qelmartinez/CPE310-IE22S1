##I. For Loop and While Loop

i.i For Loop

used when:
-you know where to end

```python
#1. For Loop

print("For loop")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print("-", fruit)
```

i.ii While Loop

used when: <br>
-you don't know what's the end

```python
#2. While Loop
print("\nWhile loop:")
count = 1
while count <= 5:
  print("Count:", count)
  count += 1
```

##II. Break and Continue

```python
for i in range(1, 6):
  if i == 3:
      continue
  print(i)
```

```python
for i in range(1, 6):
  if i == 3:
    break
  print(i)
```

continue vs break
<br> continue: skips the number that i ==
<br> break: terminates/ stops by the number that i ==
